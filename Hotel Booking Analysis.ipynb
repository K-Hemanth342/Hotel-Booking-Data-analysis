# **Cancellations and Repeated Guests**
What factors contribute the most to booking cancellations (is_canceled)?

Do repeated guests have a different cancellation rate compared to new guests?

How do previous cancellations and previous bookings that were not canceled influence future booking behavior?

What percentage of bookings are from repeated guests (is_repeated_guest), and how do their preferences differ from first-time guests?

```python
# Factors contribute the most to booking cancellations (is_canceled)?

# Group by categorical columns and calculate cancellation rate
categorical_factors = ['customer_type', 'reserved_room_type', 'deposit_type', 'market_segment']
for factor in categorical_factors:
    cancel_rate = data.groupby(factor)['is_canceled'].mean().reset_index()
    cancel_rate.columns = [factor, 'Cancellation Rate']
    print(cancel_rate)
```

```python
# numerical_factors analysis for cancellations
numerical_factors = ['lead_time', 'total_of_special_requests', 'adr']
for factor in numerical_factors:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='is_canceled', y=factor, data=data, palette='coolwarm')
    plt.title(f'{factor} vs. Cancellations')
    plt.xlabel('Is Canceled (0=No, 1=Yes)')
    plt.ylabel(factor)
    plt.show()
```

```python
# Ploting seaborn heatmap for  Correlation heatmap for cancellations
corr_data = data[['is_canceled', 'lead_time', 'total_of_special_requests', 'adr']]
corr_matrix = corr_data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation with Cancellations')
plt.show()
```

```python
# Do repeated guests have a different cancellation rate compared to new guests?
# Calculate cancellation rate for repeated and new guests
cancellation_rate = data.groupby('is_repeated_guest')['is_canceled'].mean().reset_index()
cancellation_rate.columns = ['Is Repeated Guest', 'Cancellation Rate']
cancellation_rate['Is Repeated Guest'] = cancellation_rate['Is Repeated Guest'].map({0: 'New Guest', 1: 'Repeated Guest'})

print(cancellation_rate)
```

```python
# PLoting seaborn barplot the cancellation rates for repeated and new guests.

plt.figure(figsize=(8, 6))
sns.barplot(x='Is Repeated Guest', y='Cancellation Rate', data=cancellation_rate, palette='coolwarm')
plt.title('Cancellation Rate: Repeated Guests vs. New Guests', fontsize=16)
plt.xlabel('Guest Type', fontsize=12)
plt.ylabel('Cancellation Rate', fontsize=12)
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()
```

```python
# How do previous cancellations and previous bookings that were not canceled influence future booking behavior?

# Grouping and calculating the mean cancellation rate
prev_canc = data.groupby(['previous_cancellations', 'previous_bookings_not_canceled'])['is_canceled'].mean().reset_index()
prev_canc.columns = ['previous_cancellations', 'previous_bookings_not_canceled', 'cancellation_perc']

# Visualizing with heatmap instead of barplot
plt.figure(figsize=(12, 8))
sns.heatmap(
    data=prev_canc.pivot(index='previous_bookings_not_canceled', columns='previous_cancellations', values='cancellation_perc'),
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Cancellation Rate by Previous Cancellations and Previous Bookings Not Canceled', fontsize=16)
plt.xlabel('Previous Cancellations', fontsize=12)
plt.ylabel('Previous Bookings Not Canceled', fontsize=12)
plt.show()
```
